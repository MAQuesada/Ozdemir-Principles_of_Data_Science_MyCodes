**Example the outcomes of a single dice roll that evidence the law of the large numbers**

```python
import matplotlib.pyplot as plt
import random


def random_variable_of_dice_roll():
    return random.randint(1, 6)


num_trials = range(100, 10000, 10)
avgs = []
for i in num_trials:
    trials = []
    for trial in range(1, i):
        trials.append(random_variable_of_dice_roll())
    avgs.append(sum(trials)/float(i))
plt.plot(num_trials, avgs)
plt.xlabel("number of trials")
plt.ylabel("average")
plt.grid()
```

**Observation**
* podemos ver que si lanzamos los dados unas 10 000 veces uno tras otro vemos que es muy probable que la media de las tiradas sea 3.5 que es el valor esperado de la variable aleatoria: E[X] = 1/6 *(1) + 1/6 *(2) +1/6 *(3) +1/6 *(4) +1/6 *(5) +1/6 *(6) = **3.5**
